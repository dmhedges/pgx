```python
def get_drug():
    return input('What drug? ')
```

```python
drug = get_drug()
def cpic_levels(drug=drug):
    df = pd.read_csv('/Users/david.hedges/Documents/woodahl_lab/pgx_tool/cpic/cpicPairs.csv').fillna('')
    cpic_dict = df.to_dict('records')
    #drug = get_drug()
    genesA = []
    guideA = []
    genesB = []
    guideB = []
    genesC = []
    guideC = []
    genesD = []
    guideD = []
    genesAB = []
    guideAB = []
    genesBC = []
    guideBC = []
    genesCD = []
    guideCD = []
    for i in cpic_dict:
        if i['Drug'] == drug:
            for key,val in i.items():
                if key == 'CPIC Level' and val == 'A':
                    genesA.append(i['Gene'])
                    if val != '':
                        guideA.append(i['Guideline'])
                if key == 'CPIC Level' and val == 'B':
                    genesB.append(i['Gene'])
                    if val != '':
                        guideB.append(i['Guideline'])
                if key == 'CPIC Level' and val == 'C':
                    genesC.append(i['Gene'])
                    if val != '':
                        guideC.append(i['Guideline'])
                if key == 'CPIC Level' and val == 'D':
                    genesD.append(i['Gene'])
                    if val != '':
                        guideD.append(i['Guideline'])
                if key == 'CPIC Level' and val == 'A/B':
                    genesAB.append(i['Gene'])
                    if val != '':
                        guideAB.append(i['Guideline'])
                if key == 'CPIC Level' and val == 'B/C':
                    genesBC.append(i['Gene'])
                    if val != '':
                        guideBC.append(i['Guideline'])
                if key == 'CPIC Level' and val == 'C/D':
                    genesCD.append(i['Gene'])
                    if val != '':
                        guideCD.append(i['Guideline'])
                        
    return genesA,guideA,genesB,guideB,genesC,guideC,genesD,guideD,genesAB,guideAB,genesBC,guideBC,genesCD,guideCD
```

```python
def pharmgkb_levels(drug=drug):
    df = pd.read_csv('/Users/david.hedges/Documents/woodahl_lab/pgx_tool/cpic/cpicPairs.csv').fillna('')
    cpic_dict = df.to_dict('records')
    #drug = get_drug()
    genes1A = []
    guide1A = []
    genes1B = []
    guide1B = []
    genes2A = []
    guide2A = []
    genes2B = []
    guide2B = []
    genes3 = []
    guide3 = []
    genes4 = []
    guide4 = []
    for i in cpic_dict:
        if i['Drug'] == drug:
            for key,val in i.items():
                if key == 'PharmGKB Level of Evidence' and val == '1A':
                    genes1A.append(i['Gene'])
                    if val != '':
                        guide1A.append(i['Guideline'])
                if key == 'PharmGKB Level of Evidence' and val == '1B':
                    genes1B.append(i['Gene'])
                    if val != '':
                        guide1B.append(i['Guideline'])
                if key == 'PharmGKB Level of Evidence' and val == '2A':
                    genes2A.append(i['Gene'])
                    if val != '':
                        guide2A.append(i['Guideline'])
                if key == 'PharmGKB Level of Evidence' and val == '2B':
                    genes2B.append(i['Gene'])
                    if val != '':
                        guide2B.append(i['Guideline'])
                if key == 'PharmGKB Level of Evidence' and val == '3':
                    genes3.append(i['Gene'])
                    if val != '':
                        guide3.append(i['Guideline'])
                if key == 'PharmGKB Level of Evidence' and val == '4':
                    genes4.append(i['Gene'])
                    if val != '':
                        guide4.append(i['Guideline'])

                        
    return genes1A,guide1A,genes1B,guide1B,genes2A,guide2A,genes2B,guide2B,genes3,guide3,genes4,guide4
```

```python
def cpic_results():
    drug = get_drug()
    genesA,guideA,genesB,guideB,genesC,guideC,genesD,guideD,genesAB,guideAB,genesBC,guideBC,genesCD,guideCD = cpic_levels()
    genes1A,guide1A,genes1B,guide1B,genes2A,guide2A,genes2B,guide2B,genes3,guide3,genes4,guide4 = pharmgkb_levels()
    if genesA:
        print('CPIC LEVEL A: At least one moderate or strong action (prescribing change) recommended.')
        print('Evidence is high to moderate to change prescribing.')   
        print('Problem genes:')
        for i in genesA:
            print('-' +str(i))
        print('')
        if guideA[0] != '':
            print('For more information see:')
            for i in guideA:
                print('-' +str(i))
        print('')

    if genesAB:    
        print('CPIC LEVEL A/B: At least one weak to moderate action (prescribing change) recommended.')
        print('Evidence is weak to moderate to change prescribing.')
        print('Problem genes:')   
        for i in genesAB:
            print('-' +str(i))
        print('')
        if guideAB[0] != '':
            print('For more information see:')
            for i in guideAB:
                print('-' +str(i))
        print('')
        
    if genesB:    
        print('CPIC LEVEL B: At least one optional action (prescribing change) recommended.')
        print('Evidence is weak with little conflicting data.')
        print('Problem genes: ')
        for i in genesB:
            print('-' +str(i))
        print('')
        if guideB[0] != '':
            print('For more information see:')
            for i in guideB:
                print('-' +str(i))
        print('')

    if genesBC:    
        print('CPIC LEVEL B/C: At least one optional action (prescribing change) recommended.')
        print('Evidence is weak and can vary.')
        print('Problem genes:')   
        for i in genesBC:
            print('-' +str(i))
        print('')
        if guideBC[0] != '':
            print('For more information see:')
            for i in guideBC:
                print('-' +str(i))
        print('')
        
    if genesC:    
        print('CPIC LEVEL C: No prescribing actions recommended.')
        print('Evidence is variable.')
        print('Problem genes:')
        for i in genesC:
            print('-' +str(i))
        print('')
        if guideC[0] != '':
            print('For more information see:')
            for i in guideC:
                print('-' +str(i))
        print('')

    if genesCD:    
        print('CPIC LEVEL C/D: No prescribing action recommended.')
        print('Evidence is variable.')
        print('Problem genes:')   
        for i in genesCD:
            print('-' +str(i))
        print('')
        if guideCD[0] != '':
            print('For more information see:')
            for i in guideCD:
                print('-' +str(i))
        print('')
        
    if genesD:    
        print('CPIC LEVEL D: No prescribing actions recommended.')  
        print('Evidence is weak and can vary.')
        print('Problem genes:')
        for i in genesD:
            print('-' +str(i))
        print('')
        if guideD[0] != '':
            print('For more information see:')
            for i in guideD:
                print('-' +str(i))
        print('')

    if genes1A:    
        print('PHARMGKB LEVEL 1A: Pharmacogenetic action recommended.')  
        print('Problem genes:')
        for i in genes1A:
            print('-' +str(i))
        print('')
        if guide1A[0] != '':
            print('For more information see:')
            for i in guide1A:
                print('-' +str(i))
        print('')
        
    if genes1B:    
        print('PHARMGKB LEVEL 1B: Variant-drug combination shows an association, which must be replicated.')  
        print('Problem genes:')
        for i in genes1B:
            print('-' +str(i))
        print('')
        if guide1B[0] != '':
            print('For more information see:')
            for i in guide1B:
                print('-' +str(i))
        print('')
        
    if genes2A:    
        print('PHARMGKB LEVEL 2A: Known pharmacogene involved, so functional significance is likely.')  
        print('Problem genes:')
        for i in genes2A:
            print('-' +str(i))
        print('')
        if guide2A[0] != '':
            print('For more information see:')
            for i in guide2A:
                print('-' +str(i))
        print('')
        
    if genes2B:    
        print('PHARMGKB LEVEL 2B: Variant-drug combination may show a small association that does not show statistical significance.')  
        print('Evidence is weak and can vary.')
        print('Problem genes:')
        for i in genes2B:
            print('-' +str(i))
        print('')
        if guide2B[0] != '':
            print('For more information see:')
            for i in guide2B:
                print('-' +str(i))
        print('')
        
    if genes3:    
        print('PHARMGKB LEVEL 3: Some evidence exists, but more evidence is needed.')  
        print('Problem genes:')
        for i in genes3:
            print('-' +str(i))
        print('')
        if guide3[0] != '':
            print('For more information see:')
            for i in guide3:
                print('-' +str(i))
        print('')
        
    if genes4:    
        print('PHARMGKB LEVEL 4: Only non-significant evidence exists.')  
        print('Problem genes:')
        for i in genes4:
            print('-' +str(i))
        print('')
        if guide4[0] != '':
            print('For more information see:')
            for i in guide4:
                print('-' +str(i))
        print('')
```

```python
def action_recommended():
    drug = get_drug()
    genesA,guideA,genesB,guideB,genesC,guideC,genesD,guideD,genesAB,guideAB,genesBC,guideBC,genesCD,guideCD = cpic_levels()
    genes1A,guide1A,genes1B,guide1B,genes2A,guide2A,genes2B,guide2B,genes3,guide3,genes4,guide4 = pharmgkb_levels()

    if genesA or genes1A or genes2A:
        print('Genetic Testing Strongly Recommended For:')
        for i in set(genesA+genes1A+genes2A):
            print('-'+str(i))
            
    if genesB or genesAB or genes1B or genes2B:
        print('Genetic Testing Recommended For:')
        for i in set(genesB+genesAB+genes1B+genes2B):
            print('-'+str(i))
        
    if genesC or genesBC or genes3:
        print('Genetic Testing is not Initially Recommended For:')
        for i in set(genesC+genesBC+genes3):
            print('-'+str(i))
            
    if genesD or genesCD or genes4:
        print('Genetic Testing is Most Likely Not Needed For:')
        for i in set(genesD+genesCD+genes4):
            print('-'+str(i))
```

```python
action_recommended()
```
