```python
import pandas as pd
import numpy as np
```

```python
uri = '/Users/david.hedges/Documents/woodahl_lab/pgx_tool/pharmVar/GRCh38/merged.tsv'
df = pd.read_csv(uri, sep='\t', na_filter=False, header=None)
df.columns = ['Haplotype Name', 'Gene', 'rsID', 'ReferenceSequence', 'Variant Start', 'Variant Stop',
             'Reference', 'Allele', 'Variant Allele Type']
#print(df)
df = df[['Gene','rsID','Variant Start','Variant Stop','Reference','Allele','Variant Allele Type']]
```

```python
def make_mtID(i):
    return 'mt'+str(i)
```

```python
df['mtID'] = df.index
df['mtID'] = df['mtID'].apply(make_mtID)
set(df['Gene'])
#df
```

```python
def chrNum(row):
    if row['Gene'] == 'CYP26A1' or row['Gene'] == 'CYP2C19' or row['Gene'] == 'CYP2C8' or row['Gene'] == 'CYP2C9':
        val = 'chr10'
    elif row['Gene'] == 'CYP2A13' or row['Gene'] == 'CYP2F1' or row['Gene'] == 'CYP2S1' or row['Gene'] == 'CYP4F2':
        val = 'chr19'
    elif row['Gene'] == 'CYP2D6':
        val = 'chr22'
    elif row['Gene'] == 'CYP2J2':
        val = 'chr1'
    elif row['Gene'] == 'CYP2R1':
        val = 'chr11'
    elif row['Gene'] == 'NUDT15':
        val = 'chr13'
    else:
        val = 'chr7'
    return val
```

```python
df['Chrom#'] = df.apply(chrNum, axis=1)
df
```
